````python
import pandas as pd
import os
import time
from dotenv import load_dotenv
from openai import OpenAI

# 1. 데이터 불러오기
file_path = r'C:\Users\user\Desktop\woogawooga\woogawooga_project\dataset\KorCCvi_v2.1_phishingonly.csv'
df_KKW_1 = pd.read_csv(file_path)

# 2. file_name별로 text 합치기
file_text_map = df_KKW_1.groupby('file_name')['text'].apply(lambda x: ' '.join(x)).reset_index()

# 3. 프롬프트 함수 정의
def build_prompt(text):
    return f"""
다음은 보이스피싱 대화 내용입니다.

💬 대화 내용: {text}

아래 8가지 유형 중에서 분류해 주세요.
1. 기관사칭형 (경찰, 검찰, 금융감독원, 법원, 우체국, 소방서, 교육청 등 각종 공공기관 및 정부기관)
2. 대출빙자형 (은행, 캐피탈, 저축은행, 대부업체 등 금융회사 및 대출 관련 모든 기관)
3. 메신저피싱형 (카카오톡, 문자, SNS, 이메일 등 메신저를 통해 가족·지인 사칭)
4. 택배사칭형 (CJ대한통운, 한진택배, 우체국택배 등 택배회사 및 배송 관련 사칭)
5. 투자권유형 (주식, 코인, 부동산, 펀드, 재테크 등 각종 투자 권유)
6. 가족지인사칭형 (형, 오빠, 엄마, 아빠, 언니, 누나 등 가까운 사람을 사칭)
7. 세금환급형 (국세청, 건강보험공단, 연금공단 등 세금·환급·보험금 관련 환급 사칭)
8. 콜백스미싱형 (문자·전화로 콜백을 유도하여 악성 앱 설치 또는 개인정보 탈취)

반드시 위 8가지 유형 중 하나만, 한 단어로만 답변해 주세요. (예: 대출빙자형)
"""

# 4. OpenAI API 준비
load_dotenv()
api_key = os.getenv("OPENAI_API_KEY")
if not api_key:
    raise ValueError("OPENAI_API_KEY가 .env 파일에 없거나 로딩되지 않았습니다.")
client = OpenAI(api_key=api_key)

# 5. GPT 호출 및 분류 결과 저장
phishing_type_dict = {}
total = len(file_text_map)
print(f"\n총 {total}개의 file_name 분류를 시작합니다.\n")

for idx, row in file_text_map.iterrows():
    file_name = row["file_name"]
    text = row["text"]
    prompt = build_prompt(text)

    try:
        response = client.chat.completions.create(
            model="gpt-4o",
            messages=[{"role": "user", "content": prompt}],
            temperature=0.0,
            timeout=20,
        )
        phishing_type = response.choices[0].message.content.strip()
        phishing_type_dict[file_name] = phishing_type
        print(f"[{idx+1}/{total}] {file_name} → {phishing_type}")
    except Exception as e:
        print(f"❌ Error at {file_name}: {e}")
        phishing_type_dict[file_name] = "분류불가"

    time.sleep(1.1)  # OpenAI rate limit 보호

# 6. 원본 데이터프레임에 phishing_type 매핑
df_KKW_1["phishing_type"] = df_KKW_1["file_name"].map(phishing_type_dict)

# 7. 결과 저장
df_KKW_1.to_csv("phishing_type.csv", index=False, encoding="utf-8-sig")
print("\n💾 분류 결과가 phishing_type.csv로 저장되었습니다.")
````

````python
import os
from dotenv import load_dotenv
import openai
import pandas as pd
import ast
import re

load_dotenv()
openai.api_key = os.getenv("OPENAI_API_KEY")

def create_conversation_context(texts):
    return [f"{i+1}. {text}" for i, text in enumerate(texts)]

def get_speaker_labels_improved(texts):
    numbered_texts = create_conversation_context(texts)
    conversation_text = '\n'.join(numbered_texts)

    prompt = f"""
### Task Instructions ###
The following is a real conversation between a phishing scammer and a victim.
Please accurately identify the speaker for each sentence and return it as a Python list.

### Speaker Labeling Rules ###
- Phishing scammer: 0
- Victim: 1

### Analysis Strategy ###
1. Understand the overall flow of the conversation to determine who holds the initiative.
2. Analyze the role and intent of each sentence to decide the speaker.
3. Consider linguistic features and contextual clues comprehensively.
4. Carefully analyze speaker patterns at the beginning and end of the conversation.

### Key Indicators for Phishing Scammer (0) ###
**Linguistic Features:**
- Authoritative and formal tone ("We", "You must", etc.)
- Use of professional terminology (prosecution, police, bank, etc.)
- Long sentences and complex explanations
- Detailed guidance on procedures or processes
- Threatening or pressuring expressions

**Behavior Patterns:**
- Holds the initiative in the conversation
- Asks many questions and requests information
- Tries to verify the victim's personal information
- Emphasizes urgency and applies pressure
- Explains complex procedures or systems

**Example Expressions:**
- "As a result of our verification..."
- "You must do it yourself"
- "The prosecution is investigating"
- "Your personal information has been leaked"
- "If you don't handle it right now..."

### Key Indicators for Victim (1) ###
**Linguistic Features:**
- Short and simple answers ("Yes", "No", "I don't have", "That's right", etc.)
- Uncertain expressions ("Well...", "Maybe", "I'm not sure")
- Passive and defensive tone
- Direct answers to questions

**Behavior Patterns:**
- Plays a passive role in the conversation
- Mainly responds and provides information
- Shows confusion or embarrassment
- Acts according to the scammer's requests
- Explains personal information or situations

**Example Expressions:**
- "Yes", "No", "I don't have", "I don't know"
- "I haven't done that"
- "What should I do?"
- "I'm not sure..."
- Answers related to personal information (name, address, account info, etc.)

### Contextual Analysis Guide ###
1. **Beginning of Conversation:** The scammer often initiates contact and reveals their identity.
2. **Information Request Pattern:** Scammer → asks/requests, Victim → answers/provides information
3. **Exercise of Authority:** Scammer impersonates an official institution and speaks authoritatively.
4. **Emphasis on Urgency:** Scammer uses expressions that apply time pressure.
5. **Procedure Guidance:** Scammer explains complex procedures or systems.

### Criteria for Difficult Cases ###
- If a sentence is ambiguous: consider the context before and after, and the flow of the conversation.
- For short sentences: judge whether it is a question or answer, and whether it is proactive or passive.
- During information exchange: check who is requesting and who is providing information.
- Emotional reactions: pay attention to expressions where the victim is confused or embarrassed.

### Conversation Content ###
{conversation_text}

### Output Instructions ###
- Carefully analyze each sentence based on the above indicators.
- Consider the overall flow of the conversation and the relationship between the speakers.
- Return a list whose length exactly matches the number of sentences.
- Do not include any other explanations or comments.

### Output Format ###
[0, 1, 0, 1, 0, 1, ...]
"""
    response = openai.chat.completions.create(
        model="gpt-4o",
        messages=[
            {"role": "system", "content": "당신은 피싱 사기 대화 분석 전문가입니다. 언어적 특징, 행동 패턴, 문맥적 단서를 종합적으로 분석하여 화자를 정확히 구분하는 능력을 가지고 있습니다."},
            {"role": "user", "content": prompt}
        ],
        max_tokens=1000,
        temperature=0
    )
    return response.choices[0].message.content

def parse_speaker_response(response_text):
    if not isinstance(response_text, str):
        return None
    response_text = response_text.strip()
    if response_text.startswith("```"):
        parts = response_text.split('```')
        for part in parts:
            part = part.strip()
            if part.startswith('[') and part.endswith(']'):
                response_text = part
                break
    match = re.search(r'(\[[\\d,\\s]+\\])', response_text)
    if match:
        try:
            return ast.literal_eval(match.group(1))
        except Exception:
            return None
    return None

def clean_text(text):
    if pd.isna(text):
        return ""
    return str(text).replace('\n', ' ').replace('\r', ' ').replace('"', "'").strip()

def fill_missing_speaker_gpt(
        file_path: str,
        output_path: str = None,
        chunk_size: int = 2
    ):
    """speaker_gpt 결측치만 GPT로 채워서 저장"""
    df = pd.read_csv(file_path, encoding='utf-8-sig')
    missing_mask = df['speaker_gpt'].isna()
    missing_idx = df.index[missing_mask]

    print(f"총 결측치 행 수: {len(missing_idx)}")

    # 결측치 행만 따로 처리
    for start in range(0, len(missing_idx), chunk_size):
        end = min(start + chunk_size, len(missing_idx))
        idx_chunk = missing_idx[start:end]
        # text 전처리 적용
        chunk_texts = [clean_text(t) for t in df.loc[idx_chunk, 'text'].tolist()]
        print(f"🧩 {start+1} ~ {end}번째 결측치 예측 중...")

        # 예측
        response = get_speaker_labels_improved(chunk_texts)
        labels = parse_speaker_response(response)

        # 오류 처리
        if labels is None or len(labels) != len(idx_chunk):
            print(f"❌ 오류 발생! {start+1}~{end}번째 결측치 건너뜀.")
            continue

        # 결과 반영
        df.loc[idx_chunk, 'speaker_gpt'] = labels
        print(f"✅ {end}번째 결측치까지 반영 완료.")

    # 파일 저장
    save_path = output_path if output_path else file_path
    df.to_csv(save_path, index=False, encoding='utf-8-sig')
    print(f"🎉 결측치 채운 파일 저장 완료: {save_path}")

# 사용 예시
if __name__ == "__main__":
    file_path = r'C:\Users\user\Desktop\woogawooga\woogawooga_project\dataset_create\phishing_label_type_speaker_최종.csv'
    output_path = r'C:\Users\user\Desktop\woogawooga\woogawooga_project\dataset_create\phishing_label_type_speaker_최종_filled.csv'
    fill_missing_speaker_gpt(file_path, output_path, chunk_size=2)
````

````python
input_path = r"C:\Users\user\Desktop\woogawooga\woogawooga_project\dataset_create\phishing_1301_labeled.csv"
output_path = "phishing_1301_labeled_fixed.csv"

with open(input_path, "r", encoding="utf-8-sig") as infile, \
     open(output_path, "w", encoding="utf-8-sig") as outfile:
    # 첫 줄(헤더)은 그대로 저장
    header = infile.readline()
    outfile.write(header)
    
    for line in infile:
        # 컬럼이 쉼표로 구분되어 있지만, 데이터 자체에도 쉼표가 많으므로
        # phishing_name 부분의 인덱스를 정확히 찾아야 함
        # (아래는 예시로 50번째라고 가정, 실제로는 직접 확인 필요)
        parts = line.strip().split(',')
        # phishing_name 컬럼의 시작과 끝 인덱스 찾기
        # 예를 들어, 항상 'p,h,i,s,h,i,n,g,_,1,3,0,1' 이런 식이면
        # 'p'부터 시작해서 ','로 끝나는 구간을 찾아서 합치기
        try:
            # phishing_name의 시작 인덱스 찾기 (예시: 'p'가 처음 나오는 위치)
            start_idx = parts.index('p')
            # phishing_name은 항상 13글자라면
            phishing_name = ''.join(parts[start_idx:start_idx+13]).replace(',', '')
            # 나머지 컬럼 재조립 (예시: speaker, text, phishing_name, phishing_type, speaker_gpt)
            new_parts = (
                parts[0:1] +  # speaker
                parts[1:start_idx] +  # text (쉼표 그대로)
                [phishing_name] +
                parts[start_idx+13:]  # 나머지 컬럼
            )
            outfile.write(','.join(new_parts) + '\n')
        except Exception as e:
            # 인덱스 못 찾으면 그냥 원본 라인 저장 (혹은 건너뜀)
            continue
````

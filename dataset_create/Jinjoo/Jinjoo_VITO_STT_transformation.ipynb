```python
import os
import glob
from pydub import AudioSegment, effects
import noisereduce as nr
import numpy as np

# --- 사용자 설정 (이 부분만 필요에 따라 수정하세요) ---

# 1. 원본 오디오 파일들이 있는 폴더 경로
SOURCE_FOLDER = "/content/drive/MyDrive/video_stt/video4"

# 2. 전처리된 파일들을 저장할 새 폴더 이름
#    이 폴더는 원본 폴더와 같은 위치에 생성됩니다.
DESTINATION_FOLDER_NAME = "preprocessed_audio"

# 3. 처리할 오디오 파일의 확장자 목록 (필요시 다른 확장자 추가)
TARGET_EXTENSIONS = [".wav", ".mp3", ".m4a", ".mp4"]

# --- 코드 실행 부분 (여기는 수정하지 마세요) ---

# --- 1. 경로 설정 및 폴더 생성 ---
destination_path = os.path.join(os.path.dirname(SOURCE_FOLDER), DESTINATION_FOLDER_NAME)
os.makedirs(destination_path, exist_ok=True)
print(f"전처리된 파일은 '{destination_path}' 폴더에 저장됩니다.")
print("-" * 50)

# --- 2. 파일 목록 가져오기 ---
source_files = []
for ext in TARGET_EXTENSIONS:
    source_files.extend(glob.glob(os.path.join(SOURCE_FOLDER, f"*{ext}")))

if not source_files:
    print("처리할 파일을 찾지 못했습니다. SOURCE_FOLDER 경로와 확장자를 확인해주세요.")
else:
    print(f"총 {len(source_files)}개의 파일을 처리합니다.")
    print("-" * 50)

# --- 3. 파일별 전처리 루프 ---
for i, file_path in enumerate(source_files, 1):
    filename = os.path.basename(file_path)
    # 저장될 파일명에서 확장자를 .wav로 변경 (최고 음질 유지를 위해)
    dest_filename = os.path.splitext(filename)[0] + ".wav"
    dest_file_path = os.path.join(destination_path, dest_filename)

    print(f"[{i}/{len(source_files)}] '{filename}' 파일 처리 중...")

    try:
        # --- 단계 1: 파일 불러오기 ---
        sound = AudioSegment.from_file(file_path)
        print("  - 파일 로드 완료")

        # --- 단계 2: 노이즈 제거 ---
        # noisereduce 라이브러리는 NumPy 배열로 작동합니다.
        # pydub 오디오 데이터를 NumPy 배열로 변환
        frame_rate = sound.frame_rate
        samples = np.array(sound.get_array_of_samples())

        # 노이즈 제거 수행 (배경의 '쉬-' 하는 소리 같은 고정된 노이즈에 효과적)
        reduced_noise_samples = nr.reduce_noise(y=samples, sr=frame_rate, stationary=True)

        # 처리된 NumPy 배열을 다시 pydub 오디오 데이터로 변환
        reduced_noise_sound = sound._spawn(reduced_noise_samples.tobytes())
        print("  - 노이즈 제거 완료")

        # --- 단계 3: 볼륨 정규화 ---
        # 소리가 가장 큰 부분을 기준으로 전체 볼륨을 키워 음량을 표준화합니다.
        normalized_sound = effects.normalize(reduced_noise_sound)
        print("  - 볼륨 정규화 완료")

        # --- 단계 4: 결과 저장 ---
        # 음질 손실이 없는 wav 포맷으로 저장하는 것을 권장합니다.
        normalized_sound.export(dest_file_path, format="wav")
        print(f"  - ✔️ 처리 완료 및 저장: {dest_filename}\n")

    except Exception as e:
        print(f"  - ❌ 오류 발생: {filename}, 오류: {e}\n")

print("=" * 50)
print("모든 작업이 완료되었습니다.")
```

```python
# 1. 필요한 라이브러리 설치 (가장 먼저 실행해주세요)
!pip install noisereduce pydub

# 2. 라이브러리 불러오기 (import)
# 설치가 완료되면, 아래 import 코드가 정상적으로 작동합니다.
import glob
from pydub import AudioSegment, effects
import noisereduce as nr
import numpy as np
# --- 사용자 설정 (이 부분만 필요에 따라 수정하세요) ---

# 1. 원본 오디오 파일들이 있는 폴더 경로
SOURCE_FOLDER = "/content/drive/MyDrive/골라낸파일"

# 2. 전처리된 파일들을 저장할 새 폴더 이름
#    이 폴더는 원본 폴더와 같은 위치에 생성됩니다.
DESTINATION_FOLDER_NAME = "preprocessed_audio"

# 3. 처리할 오디오 파일의 확장자 목록 (필요시 다른 확장자 추가)
TARGET_EXTENSIONS = [".wav", ".mp3", ".m4a", ".mp4"]

# --- 코드 실행 부분 (여기는 수정하지 마세요) ---

# --- 1. 경로 설정 및 폴더 생성 ---
destination_path = os.path.join(os.path.dirname(SOURCE_FOLDER), DESTINATION_FOLDER_NAME)
os.makedirs(destination_path, exist_ok=True)
print(f"전처리된 파일은 '{destination_path}' 폴더에 저장됩니다.")
print("-" * 50)

# --- 2. 파일 목록 가져오기 ---
source_files = []
for ext in TARGET_EXTENSIONS:
    source_files.extend(glob.glob(os.path.join(SOURCE_FOLDER, f"*{ext}")))

if not source_files:
    print("처리할 파일을 찾지 못했습니다. SOURCE_FOLDER 경로와 확장자를 확인해주세요.")
else:
    print(f"총 {len(source_files)}개의 파일을 처리합니다.")
    print("-" * 50)

# --- 3. 파일별 전처리 루프 ---
for i, file_path in enumerate(source_files, 1):
    filename = os.path.basename(file_path)
    # 저장될 파일명에서 확장자를 .wav로 변경 (최고 음질 유지를 위해)
    dest_filename = os.path.splitext(filename)[0] + ".wav"
    dest_file_path = os.path.join(destination_path, dest_filename)

    print(f"[{i}/{len(source_files)}] '{filename}' 파일 처리 중...")

    try:
        # --- 단계 1: 파일 불러오기 ---
        sound = AudioSegment.from_file(file_path)
        print("  - 파일 로드 완료")

        # --- 단계 2: 노이즈 제거 ---
        # noisereduce 라이브러리는 NumPy 배열로 작동합니다.
        # pydub 오디오 데이터를 NumPy 배열로 변환
        frame_rate = sound.frame_rate
        samples = np.array(sound.get_array_of_samples())

        # 노이즈 제거 수행 (배경의 '쉬-' 하는 소리 같은 고정된 노이즈에 효과적)
        reduced_noise_samples = nr.reduce_noise(y=samples, sr=frame_rate, stationary=True)

        # 처리된 NumPy 배열을 다시 pydub 오디오 데이터로 변환
        reduced_noise_sound = sound._spawn(reduced_noise_samples.tobytes())
        print("  - 노이즈 제거 완료")

        # --- 단계 3: 볼륨 정규화 ---
        # 소리가 가장 큰 부분을 기준으로 전체 볼륨을 키워 음량을 표준화합니다.
        normalized_sound = effects.normalize(reduced_noise_sound)
        print("  - 볼륨 정규화 완료")

        # --- 단계 4: 결과 저장 ---
        # 음질 손실이 없는 wav 포맷으로 저장하는 것을 권장합니다.
        normalized_sound.export(dest_file_path, format="wav")
        print(f"  - ✔️ 처리 완료 및 저장: {dest_filename}\n")

    except Exception as e:
        print(f"  - ❌ 오류 발생: {filename}, 오류: {e}\n")

print("=" * 50)
print("모든 작업이 완료되었습니다.")
```

```python
import os
import glob
import csv

# --- 1. 사용자 설정 (원본 스크립트와 동일하게 설정해주세요) ---

# 동영상/음성 파일이 있는 폴더 경로입니다.
TARGET_FOLDER = "/content/drive/MyDrive/우가우가/정부기관사칭및기타_송진주"

# 원본 스크립트가 결과를 저장한 CSV 파일 이름입니다.
OUTPUT_CSV_NAME = "vito_audio_영재1.csv"

# 원본 스크립트가 처리 대상으로 삼았던 파일의 확장자 목록입니다.
VIDEO_EXTENSIONS = [".mp4", ".mov", ".mkv", ".avi", ".m4a", ".mp3", ".wav"]


# --- 2. 로직 실행 (이 아래 부분은 수정할 필요 없습니다) ---

def find_unprocessed_files():
    """
    폴더의 전체 파일과 CSV에 기록된 파일을 비교하여 처리되지 않은 파일 목록을 찾습니다.
    """
    # 결과 CSV 파일의 전체 경로 생성
    csv_path = os.path.join(TARGET_FOLDER, OUTPUT_CSV_NAME)

    # 1. TARGET_FOLDER에서 모든 대상 파일 목록 가져오기 (Set으로 저장)
    try:
        all_source_files = set()
        for ext in VIDEO_EXTENSIONS:
            # os.path.basename을 사용해 파일 이름만 추출
            files = glob.glob(os.path.join(TARGET_FOLDER, f"*{ext}"))
            for f in files:
                all_source_files.add(os.path.basename(f))

        # 자기 자신(결과 CSV 파일)은 목록에서 제외
        if OUTPUT_CSV_NAME in all_source_files:
            all_source_files.remove(OUTPUT_CSV_NAME)

        if not all_source_files:
            print(f"⚠️  '{TARGET_FOLDER}' 폴더에서 대상 확장자를 가진 파일을 찾을 수 없습니다.")
            return

    except Exception as e:
        print(f"❌ 폴더를 읽는 중 오류가 발생했습니다: {e}")
        return

    # 2. CSV 파일에서 처리 완료된 파일 목록 가져오기 (Set으로 저장)
    processed_files = set()
    try:
        with open(csv_path, mode='r', encoding='utf-8-sig') as csvfile:
            reader = csv.reader(csvfile)
            header = next(reader)  # 헤더 건너뛰기

            # 'file_name' 컬럼이 몇 번째에 있는지 확인 (보통 첫 번째)
            try:
                filename_index = header.index("file_name")
            except ValueError:
                print(f"❌ CSV 파일 '{OUTPUT_CSV_NAME}'에 'file_name' 헤더가 없습니다.")
                return

            for row in reader:
                if row: # 빈 줄이 있는 경우 방지
                    processed_files.add(row[filename_index])

    except FileNotFoundError:
        print(f"ℹ️  결과 파일 '{OUTPUT_CSV_NAME}'을(를) 찾을 수 없습니다.")
        print("모든 파일이 처리되지 않은 것으로 간주합니다.")
        # CSV 파일이 없으면 processed_files는 비어있는 상태로 유지됩니다.
    except Exception as e:
        print(f"❌ CSV 파일을 읽는 중 오류가 발생했습니다: {e}")
        return

    # 3. 두 목록을 비교하여 처리되지 않은 파일 찾기 (차집합)
    unprocessed_files = all_source_files - processed_files

    # 4. 결과 출력
    print("-" * 60)
    print(f"'{TARGET_FOLDER}' 폴더와 '{OUTPUT_CSV_NAME}' 파일을 비교합니다.")
    print(f"🔍 총 {len(all_source_files)}개의 원본 파일 발견")
    print(f"✅ 총 {len(processed_files)}개의 처리된 파일 확인 (CSV 기준)")
    print("-" * 60)

    if not unprocessed_files:
        print("\n🎉 모든 파일이 CSV에 존재합니다. 처리되지 않은 파일이 없습니다!")
    else:
        print(f"\n🚨 총 {len(unprocessed_files)}개의 처리되지 않은 파일이 있습니다:")
        # 가나다순으로 정렬하여 출력
        for filename in sorted(list(unprocessed_files)):
            print(f"  - {filename}")

    print("-" * 60)


if __name__ == "__main__":
    find_unprocessed_files()
```

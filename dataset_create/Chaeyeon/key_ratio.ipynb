```python
import pandas as pd
# 두 CSV 불러오기
phish_df = pd.read_csv("C:/Users/user/Downloads/0708/woogawooga_project/dataset_create/Chaeyeon/top_700_okt_vp_keywords.csv")
normal_df = pd.read_csv("C:/Users/user/Downloads/0708/woogawooga_project/dataset_create/Chaeyeon/top_700_okt_nvp_keywords.csv")  # 일반 대화 키워드 CSV

# 전체 키워드 빈도표 생성
phish_word_freq = phish_df.groupby("word")["count"].sum()
normal_word_freq = normal_df.groupby("word")["count"].sum()

# 공통 키워드 기준 병합
df = pd.DataFrame({
    "phish": phish_word_freq,
    "normal": normal_word_freq
}).fillna(0)

# 총합 대비 상대 비율 계산
df["phish_ratio"] = df["phish"] / df["phish"].sum()
df["normal_ratio"] = df["normal"] / df["normal"].sum()

# 비율 차이 기반 위험도
df["risk_score"] = df["phish_ratio"] / (df["normal_ratio"] + 1e-6)

# 위험도 높은 상위 키워드  500 추출
risk_keywords = df.sort_values("risk_score", ascending=False).head(500)
```
